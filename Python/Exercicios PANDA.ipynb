```python
# 3. Selecione todas as linhas com país igual a Brasil

dados_brasil = data.loc[data["Country"]=='Brazil']
dados_brasil

```

```python
# 4. Os dados do brasil possui dados ausentes (missing data)? Se sim, selecione todas as linhas
# que contenham algum dado ausente

dados_brasil_nulos = dados_brasil[dados_brasil.isnull().any(axis=1)]
dados_brasil_nulos
```

```python
# 5. Qual o total de valores ausentes para cada coluna?

dados_brasil.isnull().sum()
```

```python
# 6. Remova linhas com pelo menos 1 elemento ausente das colunas: 'FormalEducation', 'UndergradMajor', 'Age'.

dados_brasil_limpos = dados_brasil.dropna(subset=['FormalEducation', 'UndergradMajor', 'Age'])
dados_brasil_limpos
```

```python
# 7. Crie um subconjunto de dados com as colunas: 
# Respondent, Country, Student, Employment, FormalEducation, UndergradMajor, Exercise, Age


dados_analise = dados_brasil_limpos.filter(['Respondent', 'Country', 'Student', 'Employment', 'FormalEducation', 'UndergradMajor', 'Exercise', 'Age'])
dados_analise

##print(dados_analise.info()) 
```

```python
# 8. Exiba as contagens de frequência dos valores únicos das colunas: Student, Exercise, Employment, FormalEducation, UndergradMajor, Exercise e Age

print ('          Student\n')
student_frequencia = dados_analise['Student'].value_counts();
print(student_frequencia)
print('\n')
print(student_frequencia.describe())
print('\n\n')

print ('          Exercise\n')
exercise_frequencia = dados_analise['Exercise'].value_counts();
print(exercise_frequencia)
print('\n')
print(exercise_frequencia.describe())
print('\n\n')

print ('          Employment\n')
employment_frequencia = dados_analise['Employment'].value_counts();
print(employment_frequencia)
print('\n')
print(employment_frequencia.describe())
print('\n\n')

print ('          FormalEducation\n')
formalEducation_frequencia = dados_analise['FormalEducation'].value_counts();
print(formalEducation_frequencia)
print('\n')
print(formalEducation_frequencia.describe())
print('\n\n')

print ('          UndergradMajor\n')
UndergradMajor_frequencia = dados_analise['UndergradMajor'].value_counts();
print(UndergradMajor_frequencia)
print('\n')
print(UndergradMajor_frequencia.describe())
print('\n\n')

print ('          Age\n')
Age_frequencia = dados_analise['Age'].value_counts();
print(Age_frequencia)
print('\n')
print(Age_frequencia.describe())
print('\n\n')

```
